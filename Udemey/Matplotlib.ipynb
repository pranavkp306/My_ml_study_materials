## Matplotlib

```python
from scipy.stats import norm
x=np.arange(-3,3,0.001)
plt.plot(x,norm.pdf(x))
plt.show()
```

### multiple plot on one graph

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1.0,0.5))# mean around 1, std deviation =0.5
plt.show()
```

### save it to a file

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1.0,0.5))# mean around 1, std deviation =0.5
plt.savefig("Matplot.png",dpi=300)
```

### adjust the axes

```python
axs=plt.axes()
axs.set_xlim([-5,5])
axs.set_ylim([0,2.0])
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1.0,0.5))# mean around 1, std deviation =0.5
```

### adding grid lines

```python
plt.grid()
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1.0,0.5))# mean around 1, std deviation =0.5
```

### color and line types

```python
plt.plot(x,norm.pdf(x),'r')
plt.plot(x,norm.pdf(x,1.0,0.5),'b:')# mean around 1, std deviation =0.5
```

### labelling axes and legend

```python
plt.plot(x,norm.pdf(x),)
plt.plot(x,norm.pdf(x,1.0,0.5))# mean around 1, std deviation =0.5
plt.xlabel("xlabes",size=15)
plt.ylabel("ylabes",size=15)
plt.xlim([-5,5])
plt.ylim([0,1.0])
plt.legend(["normal","mean1"],loc=2)
```

## Pie chart

```python
val=[12,55,4,32,14]
explode=[0,0,0.5,0,0]
labels=["india","usa","russia","china","europe"]
plt.figure(figsize=(16,8))
plt.pie(val,labels=labels,radius=1,explode=explode,autopct="%.2f %%")

plt.legend(loc=4)
plt.show()
```

## Bar chart

```python
val=[12,55,4,32,14]
colors=['r','g','b','c','m']
plt.bar(range(0,5),val,color=colors)
plt.show()
```

# Histogram

```python
income=np.random.normal(27000,15000,10000)
plt.hist(income,50)
plt.show()
```

# Box plot and Whisker plot

```python
uniformskewed=np.random.rand(100)*100-40
high_outliers=np.random.rand(100)* 50 +100
low_outliers=np.random.rand(100)* -50 -100
data=np.concatenate((uniformskewed,high_outliers,low_outliers))
plt.boxplot(data)

plt.show()
```
