# Seaborn
Advanced matplotlib with more function

```python
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_excel("sales.xls",sheet_name='Orders')
```

```python
df.head()
```

```python
df.info()
```

## <font color='orange'> Relational Plot


```python
# relational plot
sns.relplot('Sales','Profit',data=df)
```

```python
# relational plot adding hue, hue mustbe a categotical column
sns.relplot('Sales','Profit',data=df,hue='Order Priority')
```

```python
# relational plot adding style, style mustbe a categotical column
sns.relplot('Sales','Profit',data=df,hue='Order Priority',style='Order Priority')
```

```python
# relational plot adding style, style mustbe a categotical column
sns.relplot('Sales','Profit',data=df,hue='Order Priority',style='Ship Mode')
```

```python
# relational plot adding size, size mustbe a categotical column
sns.relplot('Sales','Profit',data=df,hue='Order Priority',style='Ship Mode',size='Discount')
```

```python
# relational plot changing size of size
sns.relplot('Sales','Profit',data=df,size='Discount',sizes=(25,300))
```

## <font color='orange'> Linechart Plot


```python
df.head()
```

```python
# simple line graph
sns.relplot('Order Date','Sales',data=df,kind='line')
```

```python
# we create a new colum to extract year from order date
df['Year']=df['Order Date'].dt.year
```

```python
# ploting sale vs year line graph
sns.relplot('Year','Sales',data=df,kind='line')
```

the shaded are is +- 95% of confidence intervel from mean (line in middle)

```python
# we can chane this interval
sns.relplot('Year','Sales',data=df,kind='line',ci=None)# if ci ='sd' then confident interval of standerd deviation
```

```python
# the graph above shows the mean of sales , if we want actual value
sns.relplot('Year','Sales',data=df,kind='line',estimator=None)
```

```python
# using hue
sns.relplot('Year','Sales',data=df,kind='line',hue='Ship Mode')
```

```python
sns.relplot('Year','Sales',data=df,kind='line',hue='Ship Mode',style="Product Category")
```

```python
# creating different categories in different columns
sns.relplot('Year','Sales',data=df,kind='line',hue='Ship Mode',col="Product Category",ci=None)
```

```python
sns.relplot('Year','Sales',data=df,kind='line',hue='Ship Mode',col="Region",ci=None)
```

```python
# since we have 8 regions its plot in single row , we can change it using col_wrap
sns.relplot('Year','Sales',data=df,kind='line',hue='Ship Mode',col="Region",ci=None,col_wrap=4)#col_wrap=4:plot 4plot in a column
```

```python
sns.relplot('Sales','Profit',data=df,col="Customer Segment",col_wrap=2)
```

```python
sns.relplot('Sales','Profit',data=df,col="Customer Segment",col_wrap=2,hue="Ship Mode")
```

```python
sns.relplot('Sales','Profit',data=df,col="Customer Segment",row="Product Category",hue="Ship Mode")
```
