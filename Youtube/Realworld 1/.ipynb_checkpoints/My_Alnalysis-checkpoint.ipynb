```python
df
```

<font color="#24e2ff"><h3>We now split the city form Puirchase address



<font color="#ff2448"><h3>
    But there are some porblems some states may have similar city name , so we must consider the states too

##### we are creating two function to split the address

```python
def get_city(address):
    return address.split(',')[1]#to get city name
def get_state(address):
    return address.split(',')[2].split(' ')[1]# to get only the state code without pin
```

```python
df['City']=df["Purchase Address"].apply(lambda x: f"{get_city(x)}({get_state(x)})")
```

```python
df.head()
```

```python
df['City'].value_counts()
```

```python
res=df.groupby('City').sum()
res
```

```python
plt.figure(figsize=(17,8))
plt.bar(res.index,res['Total_value'],color='#24ffed')
plt.title("Total sales in each city",size=20)
plt.xlabel("Cities",size=15)
plt.ylabel("Total sales in millions USD",size=20)
plt.xticks(res.index, rotation='vertical')
plt.savefig("Total sales in each city.png",dpi=300)

plt.show()
```

<font color='#93d60d'>  <h5>for that we need to convert Order Date into date time object using *pd.to_datetime()*

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df.head()
```

<h6> Adding a new column Hour and Minutes

```python
df["Hour"]=df['Order Date'].dt.hour
df["Minute"]=df['Order Date'].dt.minute
```

```python
df.head()
```

##### Plotting

```python
res=df.groupby('Hour').count()
res
```

```python
plt.figure(figsize=(17,8))
plt.plot(res.index,res,"#c309e8")
plt.title("Time vs Selling",size=20)
plt.xlabel("Time(in 24 hrs)",size=15)
plt.ylabel('Number of Orders',size=15)
plt.xticks(res.index)
plt.grid()

plt.show()
```

<font color="red"><h1>Question 4</h1></font>
<font color='#f54ef2'>  <h3>What products are most often sold together?

```python
df.head()
```

<font color='#fcba03'>creating new df with only duplicate order id

```python
df2=df[df['Order ID'].duplicated(keep=False)]#keep=False will keep all occurences of data
df2.head()
```

<font color='#fcba03'>creating new column with name grouped with products 

```python
df2['Grouped']=df2.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df2.head()
```

<font color='#fcba03'>
    Since the rows are repeated as many ties as the order id we remove the duplicate rows

```python
df2=df2[['Order ID','Grouped']].drop_duplicates()
```

```python
df2.head
```

```python
res=df2['Grouped'].value_counts()
res[:10]
```

```python
from itertools import combinations
from collections import Counter

count=Counter()
for row in df2['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list, 2)))
count.most_common(10)
```
