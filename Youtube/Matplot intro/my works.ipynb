```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#plt.plot(x,y, label="2x",c='red',linewidth=3,marker="+",markersize=15,markeredgecolor='blue',linestyle="--")
plt.plot(x,y, 'b*--', label="2x")
plt.title("Our first graph",fontdict={"fontname":'comic sans ms','fontsize':20})
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

## basic graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y)
```

#### adding title and x,y axis names

```python
plt.title("X vs 2X",fontdict={"fontname":'comic sans ms','fontsize':20})
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.plot(x,y)
plt.show()
```

#### giving values to x and y axis

```python
plt.title("X vs 2X",fontdict={'fontsize':20})
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.plot(x,y)


plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```

#### adding legend(small box showing what the line explain)

```python
plt.plot(x,y,'o',label="Twise of X",) #A 'o' after the x,y will only shows the point at each values
plt.legend()
```

#### Changing color width and type of a line
**Marker** means the point at each value of y with respect to x

```python
plt.plot(x,y, label="2X",color='red',linewidth=2,linestyle="--", marker="o",markersize=15,markeredgecolor='blue',)
plt.legend()
```

**we can do the same like color, marker, line type using the shorthand method**

```python
plt.plot(x,y, 'b+--', label="2x")    #"b+--" means blue color, + as marker and dotted line

```

## using numpy and multiple plot

```python
x2=np.arange(0,5,0.5)
x2
```

```python
plt.figure(figsize=(7,7))   # change the figsize using plt.figure           
plt.plot(x2,x2**2)           
               
plt.show()
```

```python
a=np.arange(1,15,0.5)
plt.figure(figsize=(15,5),dpi=300)

plt.subplot(1,2,1)
plt.plot(a,a**2,label="2*A")
plt.title("2*A function")
plt.legend()

plt.subplot(1,2,2)
plt.plot(a,np.log(a),'r--',label="log(A)")
plt.title("log(A) function")
plt.legend()

plt.suptitle("Main title",size=20)
plt.show()

```

**The subplots() Function**

The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.
The third argument represents the index of the current plot.

**plt.subplot(1, 2, 1)**
#the figure has 1 row, 2 columns, and this plot is the first plot.

**plt.subplot(1, 2, 2)**
#the figure has 1 row, 2 columns, and this plot is the second plot.

###  **we can even split and draw the line**

```python
plt.figure(figsize=(7,7),dpi=100)
plt.plot(a[:18],a[:18]**2,'g*-' ,label="till 18" )
plt.plot(a[17:],a[17:]**2,'r*--' ,label="after 18" )
plt.legend()
```

## saving the graph

```python
a=np.arange(1,15,0.5)
plt.figure(figsize=(15,5),dpi=300)

plt.subplot(1,2,1)
plt.plot(a,a**2,label="2*A")
plt.title("2*A function")
plt.legend()

plt.subplot(1,2,2)
plt.plot(a,np.log(a),'r--',label="log(A)")
plt.title("log(A) function")
plt.legend()

plt.suptitle("Main title",size=20)


plt.savefig("my graph.png",dpi=300)

plt.show()# it should come last 
```
