```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gas=pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,5))
plt.title('USA vs Canada Gas prices',fontdict={'fontsize' : 20})
plt.plot(gas.Year,gas.USA,'ro-',label='USA')
plt.plot(gas.Year,gas.Canada,'bo-', label='Canada')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Dollars/Gallon')
plt.xticks(gas.Year[::3].tolist()+[2011])

plt.savefig('USA vs Canada Gas prices.png',dpi=300)
plt.show()
```

# fifa data

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head()
```

## Histogram

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,color='#f5bc42')
plt.ylabel('Number of players')
plt.xlabel('Skill level')
plt.title('Distribution of players skill in fifa 2018')



plt.xticks(bins)
plt.show()
```

# Pies chart

```python
fifa['Preferred Foot']
```

```python
fifa['Preferred Foot'].value_counts()
```

```python
right=fifa['Preferred Foot'].value_counts()[0]
left=fifa['Preferred Foot'].value_counts()[1]
```

```python
labels=["Right","Left"]
colors=['#1ee3d2','#ffeb0d']
plt.pie([right,left],labels=labels,colors=colors,startangle = 90,shadow = True, autopct='%.2f %%', radius=2)

plt.legend(title = "Preferred Foot")
plt.show()
```

## Pie chart 2

```python
fifa["Weight"]=[int(x.strip('lbs')) if type(x)==str else x for x in fifa["Weight"]]
fifa["Weight"].isnull().value_counts()
```

```python
light=fifa.loc[fifa["Weight"]<125].count()[0]
light_medium=fifa.loc[(fifa["Weight"]>=125) & (fifa["Weight"]<150)].count()[0]
medium=fifa.loc[(fifa["Weight"]>=150) & (fifa["Weight"]<175)].count()[0]
medium_heavy=fifa.loc[(fifa["Weight"]>=175) & (fifa["Weight"]<200)].count()[0]
heavy=fifa.loc[fifa["Weight"]>=200].count()[0]

plt.figure(figsize=(10,10))
plt.style.use('ggplot')
weights=[light, light_medium, medium, medium_heavy ,heavy]
lablels=['Under 125', '125-150', '150-175', '175-200' ,'Above 200']
explode=(.8,.2,0,0,.4)
plt.title('Weight distribution of fifa players (in lbs)',size=20)


plt.pie(weights,labels=lablels,radius=0.8,autopct='%.2f %%',pctdistance=.9,explode=explode)
plt.legend(loc = 'lower left',title='Weight distribution')

plt.savefig("Weight distribution of fifa players",dpi=300)
plt.show()
```

# Box plot

```python
plt.style.use('default')
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
Manchester=fifa.loc[fifa.Club=='Manchester City']['Overall']
Cambridge=fifa.loc[fifa.Club=='Cambridge United']['Overall']
labels=['barcelona','Manchester','Cambridge']
plt.figure(figsize=(5,8))



boxes=plt.boxplot([barcelona,Manchester,Cambridge],labels=labels, patch_artist=True)

for box in boxes["boxes"]:
    #changing box colors
    box.set(color='#1ce1ff',linewidth=1.2)
    
    #changing fill color !make  patch_artist=True in plot section
    box.set(facecolor='#c9c5c5')

plt.title('Football teams comparison',size=20)
plt.ylabel('FIFA over all rating')
plt.show()
```
