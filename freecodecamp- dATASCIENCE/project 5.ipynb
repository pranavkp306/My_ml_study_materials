```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
%matplotlib inline
```

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df.tail()
```

```python
df.info()
```

```python
x = df['Year']
y = df['CSIRO Adjusted Sea Level']
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('CSIRO Adjusted Sea Level')
```

```python
res = linregress(x,y)
```

```python
res
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title("Rise in Sea Level")

x2 = np.arange(1880, 2050)
y2=[(res.intercept + res.slope*x) for x in x2]
plt.plot(x2, y2, 'r', label='fitted line 1')




plt.show()
```

```python
df2=df.loc[df['Year'] >= 2000]
x=df2['Year']
y=df2['CSIRO Adjusted Sea Level']
res2=linregress(x,y)
res2
```

```python
x = df['Year']
y = df['CSIRO Adjusted Sea Level']

plt.figure(figsize=(10,6))
plt.scatter(x,y,label='original points')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title("Rise in Sea Level")



res = linregress(x,y)

x2 = np.arange(1880, 2050)
y2=[(res.intercept + res.slope*x) for x in x2]
plt.plot(x2, y2, 'r', label='fitted line 1')






df2=df.loc[df['Year'] >= 2000]
x2=df2['Year']
y2=df2['CSIRO Adjusted Sea Level']
res2=linregress(x2,y2)


x3 = np.arange(2000, 2050)
y3=[(res2.intercept + res2.slope*x) for x in x3]
plt.plot(x3, y3, 'g', label='fitted line 2')

plt.legend()
plt.show()
```
