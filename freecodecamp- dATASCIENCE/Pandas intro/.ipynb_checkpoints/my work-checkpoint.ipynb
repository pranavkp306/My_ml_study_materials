```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("btc-market-price.csv",header=None,names=['Timestamp','Price'],index_col=0,parse_dates=True)
df.head()
```

```python
df.plot(figsize=(10,5))
```

```python
df2=pd.read_csv("eth-price.csv",index_col=0,parse_dates=True)
df2.head()
```

```python
df2["Value"].plot(figsize=(10,5))
```

```python
compare=pd.DataFrame(index=df.index)
compare.head()
```

```python
compare["Bitcoin"]=df['Price']
compare["Ether"]=df2['Value']
compare.head()
```

```python
compare.plot(figsize=(12,6))
```

```python
compare.loc["2017-12-01":"2018-01-01"].plot(figsize=(12,6))
```
