```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('fcc-forum-pageviews.csv')
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.index=df['date']
```

```python
df=df.drop(columns='date')
```

```python
df
```

```python
df=df[(df['value']>=df['value'].quantile(0.025)) & (df['value']<=df['value'].quantile(0.975))]
```

```python
df
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'r')
plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
plt.xlabel('Date')
plt.ylabel("Page Views")
```

```python
df_bar=df.copy()
```

```python
df_bar['Year'] = df.index.year
df_bar['Month'] = df.index.month
df_bar_ave = df_bar.groupby(['Year', 'Month']).mean().unstack()
df_bar_ave
```

```python
months=['January', 'February', 'March', 'April' , 'May' , 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_bar_ave.plot(kind='bar',figsize=(15,10)).figure
plt.xlabel("Years")
plt.ylabel("Average Page Views")
plt.legend(fontsize =10, labels=months,title='Months')
```

```python
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]
```

```python
df_box.info()
```

```python
fig, axes = plt.subplots(figsize=(20, 5), ncols=2)
box_plot_year = sns.boxplot(x=df_box['year'], y=df_box.value, ax=axes[0])
box_plot_year.set_title("Year-wise Box Plot (Trend)")
box_plot_year.set_xlabel('Year')
box_plot_year.set_ylabel('Page Views')

order=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
box_plot_month = sns.boxplot(x=df_box['month'], y=df_box.value,order=order,ax=axes[1])
box_plot_month.set_title("Month-wise Box Plot (Seasonality)")
box_plot_month.set_xlabel('Month')
box_plot_month.set_ylabel('Page Views')   
```
