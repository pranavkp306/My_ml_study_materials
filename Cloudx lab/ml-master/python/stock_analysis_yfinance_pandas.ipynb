We will now write a function to calculate returns for a specfied period. The function will take a datafram and number of days as input and return a pandas series which gives the return for each item after the specified days.

```python
def calculate_series_return(df, days): 
    returns=[]
    dates_int=[]
    close=df['Close']
    dates=df.index
    length=len(close)-days
    for i in range(0, length):
        returns.append(close[i+days]/close[i]-1)
        dates_int.append(dates[i])
    return pd.Series(returns, index=dates_int)
```

We will now calculate the returns for 1, 2, 5 (1 wk), 10 (2 wk), 21(1 month), 250(1 yr), 500(2 yr), 750(3 yr), 1250(5 years), 2500(10 years) days. Note that there 5 working days in a week. Approximately 21 days in a months and 250 days in a year. 

```python
data['1_day']=calculate_series_return(data, 1)
data['2_day']=calculate_series_return(data, 2)
data['5_day']=calculate_series_return(data, 5)
data['10_day']=calculate_series_return(data, 10)
data['1_month']=calculate_series_return(data, 21)
data['1_year']=calculate_series_return(data, 250)
data['2_year']=calculate_series_return(data, 500)
data['3_year']=calculate_series_return(data, 750)
data['5_year']=calculate_series_return(data, 1230)
data['10_year']=calculate_series_return(data, 2500)
```
