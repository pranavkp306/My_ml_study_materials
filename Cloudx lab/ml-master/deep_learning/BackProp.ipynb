## Here we cover back propagation


```python
import numpy as np
```

```python
## We are given x and y
x_raw = np.array([1,2,3,4, 5]).reshape(-1,1)
m = x_raw.shape[0]
ones = np.ones(m).reshape(-1, 1)
x = np.concatenate((ones, x_raw), axis=1)

y = (x_raw*x_raw).reshape(1,-1)[0]

print("x:")
print(x)
print("----")
print("y:")
print(y)
```

```python
## Assume values of w
w1 = np.random.random(2)
w2 = np.random.random(2)
print(w1)
print(w2)
```

## Lets Calculate h1

```python
h1 = x.dot(w1)
```

```python
h1
```

## Lets Calculate o1

```python
h1_bias = np.concatenate((ones, h1.reshape(-1,1)), axis=1)
```

```python
h1_bias
```

```python
o1 = h1_bias.dot(w2)
```

```python
o1
```

```python
y
```

# Lets Compute error

```python
diff = o1 - y
```

```python
diff
```

```python
error = np.sum(diff*diff)/m
```
